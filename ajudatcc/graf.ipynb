```python
import pandas as pd 
import matplotlib as plt
import matplotlib.pyplot as plt

plt.figure()
```

```python
abun = pd.read_excel("abun.xlsx")

j= 0
familia = []
count = []	

for i in abun["subfamilia"]:

    x = i
    y = abun["qnt"][j]
    j += 1
    familia.append(x)
    count.append(y)
    
plt.figure(figsize=(13, 3))
print(count)
b = 'blue'
g = 'deeppink'
r = 'red'
c =  'cyan'
m = 'magenta'
y = 'green'
k = 'brown'
w = 'orange'

lista = [b, g, r, c, m, y, k, w]

for i in range(len(familia)):
    plt.bar(familia[i], count[i], width=.2, color = lista[i])
plt.yticks([0, 50, 100, 150, 200, 250, 300, 350,374])
plt.grid(False)
plt.title("Abundância das Subfamílias de Stratiomyidae")
plt.ylabel("Número de espécimes")

plt.show()

```
